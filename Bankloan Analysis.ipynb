```python
sns.histplot(df['age'], kde=True, bins=20)
plt.title('Age Distribution')
plt.show()
```

### ðŸ“Š Age Distribution

This graph shows how customer ages are spread across the dataset.  
Most customers fall within a common age range, while very young and very old ages appear less frequently.


```python
sns.scatterplot(x='age', y='monthly_income', hue='loan_approved', data=df, alpha=0.7)
plt.title(' Age vs Income')
plt.show()
```

### ðŸ“ˆ Age vs Monthly Income

This scatter plot shows the relationship between customer age and monthly income.  
The color distinction highlights loan approval status, indicating how income and age together influence loan approval patterns.


```python
grouped = df.groupby("property_ownership")["loan_approved"].value_counts().unstack()

grouped.plot(kind="bar", color=["#fc8d62", "#66c2a5"])
plt.title("Loan Approval by Property Ownership")
plt.xlabel("Property Ownership")
plt.ylabel("Count")
plt.show()
```

### ðŸ“Š Loan Approval by Property Ownership

This bar chart compares loan approval status across different property ownership types.  
It shows that customers with certain property ownership categories tend to have higher loan approval counts than others.


```python
values = df["loan_approved"].value_counts()

plt.pie(
    values,
    labels=values.index,
    autopct="%1.1f%%",
    colors=["#66c2a5", "#fc8d62"],
    explode=(0.05, 0),   # ðŸ‘ˆ explode effect
    startangle=90
)
plt.title("Loan Approval Percentage")
plt.show()
```

### ðŸ¥§ Loan Approval Percentage

This pie chart shows the proportion of approved and rejected loans in the dataset.  
It highlights the overall loan approval rate, making it easy to compare approval versus rejection at a glance.


```python
sns.boxplot(
    x="loan_approved",
    y="age",
    hue="gender",
    data=df,
    palette="Set1"
)
plt.title("Age Distribution by Loan Approval and Gender")
plt.xlabel("Loan Approved")
plt.ylabel("Age")
plt.show()
```

### ðŸ“¦ Age Distribution by Loan Approval and Gender

This box plot compares the age distribution of customers based on loan approval status and gender.  
It helps identify differences in age patterns and highlights potential outliers between approved and rejected loans.


```python
sns.violinplot(
    x="loan_approved",
    y="age",
    data=df,
    palette="Set2"
)
plt.title("Age Distribution by Loan Approval")
plt.xlabel("Loan Approved")
plt.ylabel("Age")
plt.show()
```

### ðŸŽ» Age Distribution by Loan Approval

This violin plot shows how customer ages are distributed for approved and rejected loans.  
It combines distribution shape and density, making it easy to compare age patterns between the two groups.


```python
sns.countplot(x="gender", hue="loan_approved", data=df)
plt.title("Loan Approval by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

### ðŸ“Š Loan Approval by Gender

This count plot shows the number of approved and rejected loans for each gender.  
It helps compare loan approval patterns between genders at a glance.


```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap='coolwarm', annot=False) 
plt.title('9. Heatmap: Correlation Matrix')
plt.show()
```

### ðŸ”¥ Correlation Heatmap

This heatmap shows the correlation between numerical variables in the dataset.  
It helps identify strong positive or negative relationships and understand how features are related to each other.
